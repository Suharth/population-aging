```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel('median_age.xlsx',skiprows=15,usecols="H:V")
plt.figure(figsize=(10,10))
font1={'family':'serif','color':'blue','size':25}
font2={'family':'serif','color':'darkred','size':20}
x=data.loc[0]
y=data.loc[1]
plt.plot(x,y,marker='o')
y=data.loc[3]
plt.plot(x,y,marker='o')
y=data.loc[5]
plt.plot(x,y,marker='o')
plt.legend(["WORLD","More Developed Regions","Least Developed Countries"])
plt.xlabel("Year",fontdict=font2)
plt.ylabel("Median age",fontdict=font2)
plt.xticks(rotation=-60,fontsize=15)
plt.yticks(fontsize=15)
plt.ylim(17,44)
plt.title("Population Aging",fontdict=font1)
plt.savefig('population_aging.jpg')
```
